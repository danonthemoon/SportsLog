```python
def boxScoreScrape(team,date):
    bat_data = pd.DataFrame()
    pitch_data = pd.DataFrame()
    gameDate = date
    homeTeam=team
    if homeTeam in teams.keys():
        homeTeam = teams[homeTeam]
        
    url = 'https://www.baseball-reference.com/boxes/%s/%s%s0.shtml' % (homeTeam,homeTeam,gameDate)
    
    # hacky way to account for the doubleheader game i saw
    if gameDate == '20220802':
        url = 'https://www.baseball-reference.com/boxes/%s/%s%s2.shtml' % (homeTeam,homeTeam,gameDate)
    
    
    res = requests.get(url)    
    comm = re.compile("<!--|-->")
    soup = BeautifulSoup(comm.sub("", res.text), 'html.parser')

    
    scorebox = soup.find(attrs = {'class' : 'scorebox'})
    scores = scorebox.findAll(attrs = {'class' : 'score'})
    awayTeamScore = int(scores[0].getText())
    homeTeamScore = int(scores[1].getText())
    
    try: 
        teamLinks=scorebox.select("a[href*=teams]")
    except:
        print(hteam,gameDate)
        return (None,None)
        
    ateam = (teamLinks[0]['href'].split('/')[2])
    awayTeam=ateam
    if awayTeam in teams.keys():
        awayTeam = teams[awayTeam]

    #################
    # game info #
    #################
        
    gameMetadata = soup.find(attrs = {'class' : 'scorebox_meta'})
    gameMetadataList = gameMetadata.find_all('div')
    gDate = gameMetadataList[0].text
    startTime = gameMetadataList[1].text.split()[2] + ' ' + gameMetadataList[1].text.split()[3]
    attendance = gameMetadataList[2].text.split()[1]
    stadium = gameMetadataList[3].text.split()[1:]
    stadium = ' '.join(str(x) for x in stadium)
    gameDur = gameMetadataList[4].text.split()[2]

    gameInfoList = [gDate, team, homeTeamScore, awayTeamScore, ateam, startTime, attendance, stadium, gameDur]

    # Create the pandas DataFrame
    game_data = pd.DataFrame([gameInfoList],columns=['Date', 'Home', 'H', 'A', 'Away', 'Time', 'Attendance', 'Stadium', 'Duration'])
    game_data['Date'] = pd.to_datetime(game_data['Date'])
        
    #################
    # home batting #
    #################


    homeBattingTable = soup.find_all("table", {"id":"%s" % battingNames[homeTeam]})
    try:
        data_rows = homeBattingTable[0].findAll('tr')
    except:
        print(homeTeam, gameDate)
        return (None,None,None)
    data_header = homeBattingTable[0].findAll('thead')
    data_header = data_header[0].findAll("tr")
    data_header = data_header[0].findAll("th")
    homeBattingRows = [[td.getText() for td in data_rows[i].findAll(['th','td'])]
        for i in range(len(data_rows))
        ]
    homeBattingStats = pd.DataFrame(homeBattingRows)
    header = []
    for i in range(len(homeBattingStats.columns)):
        header.append(data_header[i].getText())
    homeBattingStats.columns = header
    homeBattingStats = homeBattingStats.loc[homeBattingStats[header[0]] != header[0]].reset_index(drop = True)
    homeBattingStats = homeBattingStats[homeBattingStats.PA != ''].reset_index(drop = True)
    
    # adding HR stats
    homeBattingStats.insert(4,'HR',0)
    for i in range (len(homeBattingStats)):
        # Sum HRs for Team Totals row
        if i == len(homeBattingStats)-1:
            homeBattingStats.at[i,'HR']=homeBattingStats['HR'].sum()
        else:
            for detail in homeBattingStats['Details'][i].split(','):
                if 'HR' in detail:
                    if len(detail.split('·')) > 1:
                        homeBattingStats.at[i,'HR']=int(detail.split('·')[0])
                    else:
                        homeBattingStats.at[i,'HR']=1
                    break
    
    homeBattingStats.insert(1,'Pos','N/A')

    for i in range (len(homeBattingStats)-1):
        words = homeBattingStats['Batting'][i].split()
        position = words[-1]
        name = ' '.join(str(x) for x in words[:-1])
    
        homeBattingStats.at[i, 'Pos'] = position
        homeBattingStats.at[i, 'Batting'] = name
        homeBattingStats.at[i,'Team'] = homeTeam
        homeBattingStats.at[i,'Opp'] = awayTeam
        homeBattingStats.at[i,'Date'] = gameDate
        homeBattingStats.at[i,'G'] = 1
        
    homeBattingStats.drop(columns=['BA','OBP','SLG','OPS','Pit','Str','WPA','WPA+','WPA-','cWPA','aLI','acLI','RE24','PO','A'],inplace=True)

    #################
    # away batting #
    #################

    awayBattingTable = soup.find_all("table", {"id":"%s" % battingNames[awayTeam]})
    try:
        data_rows = awayBattingTable[0].findAll('tr')
    except:
        print(homeTeam, awayTeam, gameDate)
        return (None,None,None)
    data_header = awayBattingTable[0].findAll('thead')
    data_header = data_header[0].findAll("tr")
    data_header = data_header[0].findAll("th")
    awayBattingRows = [[td.getText() for td in data_rows[i].findAll(['th','td'])]
        for i in range(len(data_rows))
        ]
    awayBattingStats = pd.DataFrame(awayBattingRows)
    header = []
    for i in range(len(awayBattingStats.columns)):
        header.append(data_header[i].getText())
    awayBattingStats.columns = header
    awayBattingStats = awayBattingStats.loc[awayBattingStats[header[0]] != header[0]].reset_index(drop = True)
    awayBattingStats = awayBattingStats[awayBattingStats.PA != ''].reset_index(drop = True)
    
    # adding HR stats
    awayBattingStats.insert(4,'HR',0)
    for i in range (len(awayBattingStats)):
        # Sum HRs for Team Totals row
        if i == len(awayBattingStats)-1:
            awayBattingStats.at[i,'HR']=awayBattingStats['HR'].sum()
        else:
            for detail in awayBattingStats['Details'][i].split(','):
                if 'HR' in detail:
                    if len(detail.split('·')) > 1:
                        awayBattingStats.at[i,'HR']=int(detail.split('·')[0])
                    else:
                        awayBattingStats.at[i,'HR']=1
                    break
        
    awayBattingStats.insert(1,'Pos','N/A')

    for i in range (len(awayBattingStats)-1):
        words = awayBattingStats['Batting'][i].split()
        position = words[-1]
        name = ' '.join(str(x) for x in words[:-1])
    
        awayBattingStats.at[i, 'Pos'] = position
        awayBattingStats.at[i, 'Batting'] = name
        awayBattingStats.at[i,'Team'] = awayTeam
        awayBattingStats.at[i,'Opp'] = homeTeam
        awayBattingStats.at[i,'Date'] = gameDate
        awayBattingStats.at[i,'G'] = 1
        
    awayBattingStats.drop(columns=['BA','OBP','SLG','OPS','Pit','Str','WPA','WPA+','WPA-','cWPA','aLI','acLI','RE24','PO','A'],inplace=True)

   
    #################
    # home pitching #
    #################

    homePitchingTable = soup.find_all("table", {"id":"%s" % pitchingNames[homeTeam]})
    data_rows = homePitchingTable[0].findAll('tr')
    data_header = homePitchingTable[0].findAll('thead')
    data_header = data_header[0].findAll("tr")
    data_header = data_header[0].findAll("th")
    homePitchingRows = [[td.getText() for td in data_rows[i].findAll(['th','td'])]
        for i in range(len(data_rows))
        ]
    homePitchingStats = pd.DataFrame(homePitchingRows)
    header = []
    for i in range(len(homePitchingStats.columns)):
        header.append(data_header[i].getText())
    homePitchingStats.columns = header
    homePitchingStats = homePitchingStats.loc[homePitchingStats[header[0]] != header[0]]
    homePitchingStats.drop(columns=['Ctct','StS','StL','GB','FB','LD','Unk','IR','IS','WPA','aLI','cWPA','acLI','RE24'],inplace=True)
    homePitchingStats = homePitchingStats.reset_index(drop = True)
    for i in range (len(homePitchingStats)):
        homePitchingStats.at[i, 'Pitching'] = homePitchingStats['Pitching'][i].split(',')[0]
        homePitchingStats.at[i,'Team'] = homeTeam
        homePitchingStats.at[i,'Opp'] = awayTeam
        homePitchingStats.at[i,'Date'] = gameDate
        homePitchingStats.at[i,'G'] = 1

    #################
    # away pitching #
    #################

    awayPitchingTable = soup.find_all("table", {"id":"%s" % pitchingNames[awayTeam]})
    data_rows = awayPitchingTable[0].findAll('tr')
    data_header = awayPitchingTable[0].findAll('thead')
    data_header = data_header[0].findAll("tr")
    data_header = data_header[0].findAll("th")
    awayPitchingRows = [[td.getText() for td in data_rows[i].findAll(['th','td'])]
        for i in range(len(data_rows))
        ]
    awayPitchingStats = pd.DataFrame(awayPitchingRows)
    header = []
    for i in range(len(awayPitchingStats.columns)):
        header.append(data_header[i].getText())
    awayPitchingStats.columns = header
    awayPitchingStats = awayPitchingStats.loc[awayPitchingStats[header[0]] != header[0]]
    awayPitchingStats.drop(columns=['Ctct','StS','StL','GB','FB','LD','Unk','IR','IS','WPA','aLI','cWPA','acLI','RE24'],inplace=True)
    awayPitchingStats = awayPitchingStats.reset_index(drop = True)
    for i in range (len(awayPitchingStats)):
        awayPitchingStats.at[i, 'Pitching'] = awayPitchingStats['Pitching'][i].split(',')[0]
        awayPitchingStats.at[i,'Team'] = awayTeam
        awayPitchingStats.at[i,'Opp'] = homeTeam
        awayPitchingStats.at[i,'Date'] = gameDate
        awayPitchingStats.at[i,'G'] = 1
        
        
    bat_data = pd.concat([bat_data,homeBattingStats], ignore_index=True)
    bat_data = pd.concat([bat_data,awayBattingStats], ignore_index=True)
    bat_data = bat_data[bat_data.Batting != 'Team Totals']
    bat_data['Date'] = pd.to_datetime(bat_data['Date'])
    
    pitch_data = pd.concat([pitch_data,homePitchingStats], ignore_index=True)
    pitch_data = pd.concat([pitch_data,awayPitchingStats], ignore_index=True)
    pitch_data = pitch_data[pitch_data.Pitching != 'Team Totals']
    pitch_data['Date'] = pd.to_datetime(pitch_data['Date'])
    
    
    return (bat_data, pitch_data, game_data)
```

```python
def getAllStats(gameList):
    batting_data = pd.DataFrame()
    pitching_data = pd.DataFrame()
    game_data = pd.DataFrame()
    games = gameList
    
    i=1
    for game in games:
        homeTeam = game[0]
        date = game[1]
        gamebatting, gamepitching, gameinfo = boxScoreScrape(homeTeam,date)
        # space requests to stay out of SportsRef jail
        time.sleep(4)
        if not gamebatting.empty:
            batting_data = pd.concat([batting_data,gamebatting], ignore_index=True)
            pitching_data = pd.concat([pitching_data,gamepitching], ignore_index=True)
            game_data = pd.concat([game_data,gameinfo], ignore_index=True)
            print("logged game %i - %s %s" %(i,homeTeam, date))
        else:
            print("issue with %s %s" %(homeTeam, date))
        i+=1
    return (batting_data, pitching_data, game_data)
```

```python
# Aggregated stats for unique batters and pitchers

#can add Team to groupby for breakdown
hitters = batting.groupby(["Batting"]).sum(numeric_only=True).reset_index()
hitters.insert(4,'AVG','N/A')
for i in range(len(hitters)):
    if hitters.at[i,'H'] > 0:
        hitters.at[i,'AVG'] = round(hitters.at[i,'H'] / hitters.at[i,'AB'],3)
    else:
        hitters.at[i,'AVG'] = .000
    
hitters = hitters[hitters['PA']>0]


# can add Team to groupby for breakdown
tossers = pitching.groupby(["Pitching"]).sum(numeric_only=True).reset_index()
```

```python
# some general accumulated stats
def genstats(gamelist,hitlist,pitchlist):
    games = gamelist
    hitters = hitlist
    tossers = pitchlist
    
    numGames = len(games)

    uniqueBatters = len(hitters)
    uniquePitchers = len(tossers)

    hits = hitters['H'].sum()
    if hits != tossers['H'].sum():
        print('hit count off')

    atbats = hitters['AB'].sum()

    avg = round(hits/atbats,3)

    runs = hitters['R'].sum()
    if runs != tossers['R'].sum():
        print('run count off')

    homeruns = hitters['HR'].sum()
    if homeruns != tossers['HR'].sum():
        print('homerun count off')

    strikeouts = hitters['SO'].sum()
    if strikeouts != tossers['SO'].sum():
        print('strikeout count off')

    walks = hitters['BB'].sum()
    if walks != tossers['BB'].sum():
        print('walk count off')

    pitches = tossers['Pit'].sum()
    strikes = tossers['Str'].sum()

    print("%i Games" % numGames)
    print("%i Unique Batters" % uniqueBatters)
    print("%i Unique Batters" % uniquePitchers)
    print("%i Hits" % hits)
    print("%i ABs" % atbats)
    print("%s Batting Average" % str(avg))
    print("%i Runs" % runs)
    print("%i HRs" % homeruns)
    print("%i Ks" % strikeouts)
    print("%i BBs" % walks)
    print("%i Pitches" % pitches)
    print("%i Strikes" % strikes)
```
