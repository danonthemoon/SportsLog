```python
goat = stats[stats['Player'] == 'Russell Westbrook']
goat.head(11)
```

```python
goat.columns
```

```python
goat = goat.drop(columns = ['Unnamed: 0', 'Age', 'Tm', 'GS', 'blanl', 'blank2', 'WS/48'])
```

```python
russ = goat.drop(columns = ['Pos', 'Player', 'PER', 'USG%', 'MP','TS%', '3PAr', 'FTr', 'ORB%', 'DRB%', 'TRB%', 'AST%', 'STL%',
       'BLK%', 'TOV%', 'OWS', 'DWS', 'WS',
       'OBPM', 'DBPM', 'BPM', 'VORP',
       '2P', '2PA', '2P%', 'PF', 'DRB', 'ORB', 'eFG%'])
```

```python
russ.head(15)
```

```python
russ = russ.append({'Year':2018, 'G':80, 'FG': 757,
                                 'FGA':1687,'FG%':.449, '3P': 97, '3PA':326, '3P%':.298,
                                  'FT':417, 'FTA':556, 'FT%':.737, 
                                  'TRB':804, 'AST':820, 'STL':147, 'BLK':20, 'TOV':381, 'PTS':2028,}, ignore_index=True)

russ = russ.append({'Year':2019,'G':73, 'FG': 630,
                                 'FGA':1473,'FG%':.428, '3P': 119, '3PA':411, '3P%':.290,
                                 'FT':296, 'FTA':451, 'FT%':.656, 
                                  'TRB':807, 'AST':784, 'STL':142, 'BLK':33, 'TOV':325, 'PTS':1675,}, ignore_index=True)
```

```python
russ.head(15)
```

```python
russ["PPG"] = (russ["PTS"] / russ["G"]).round(1)
russ["APG"] = (russ["AST"] / russ["G"]).round(1)
russ["RPG"] = (russ["TRB"] / russ["G"]).round(1)
```

```python
russ.head(15)
```

```python
russ.plot('Year', 'PPG', figsize = (9,6), legend=False, grid=True);
plt.xticks(russ['Year']);
plt.xlabel('Year', fontsize=16)
plt.ylabel('Points per Game', fontsize=16)
plt.title('PPG by Year', fontsize=20)

russ.plot('Year', {'APG', 'RPG'}, figsize = (9,6), grid=True);
plt.xticks(russ['Year']);
plt.xlabel('Year', fontsize=16)
plt.ylabel('Assists & Rebounds per Game', fontsize=16)
plt.title('APG & RPG by Year', fontsize=20);
```
